```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import pandas as pd
import seaborn as sns
```

```python
# Wczytanie danych z Dataframe
df1 = pd.read_csv('Brickset-Sets-2000.csv', sep = ',')
df2 = pd.read_csv('Brickset-Sets-2001.csv', sep = ',')
df3 = pd.read_csv('Brickset-Sets-2002.csv', sep = ',')
df4 = pd.read_csv('Brickset-Sets-2003.csv', sep = ',')
df5 = pd.read_csv('Brickset-Sets-2004.csv', sep = ',')
df6 = pd.read_csv('Brickset-Sets-2005.csv', sep = ',')
df7 = pd.read_csv('Brickset-Sets-2006.csv', sep = ',')
df8 = pd.read_csv('Brickset-Sets-2007.csv', sep = ',')
df9 = pd.read_csv('Brickset-Sets-2008.csv', sep = ',')
df10 = pd.read_csv('Brickset-Sets-2009.csv', sep = ',')
df11 = pd.read_csv('Brickset-Sets-2010.csv', sep = ',')
df12 = pd.read_csv('Brickset-Sets-2011.csv', sep = ',')
df13 = pd.read_csv('Brickset-Sets-2012.csv', sep = ',')
df14 = pd.read_csv('Brickset-Sets-2013.csv', sep = ',')
df15 = pd.read_csv('Brickset-Sets-2014.csv', sep = ',')
df16 = pd.read_csv('Brickset-Sets-2015.csv', sep = ',')
df17 = pd.read_csv('Brickset-Sets-2016.csv', sep = ',')
df18 = pd.read_csv('Brickset-Sets-2017.csv', sep = ',')
df19 = pd.read_csv('Brickset-Sets-2018.csv', sep = ',')
df20 = pd.read_csv('Brickset-Sets-2019.csv', sep = ',')
df21 = pd.read_csv('Brickset-Sets-2020.csv', sep = ',')
df22 = pd.read_csv('Brickset-Sets-2021.csv', sep = ',')
df23 = pd.read_csv('Brickset-Sets-2022.csv', sep = ',')
df24 = pd.read_csv('Brickset-Sets-2023.csv', sep = ',')

dane = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21, df22, df23, df24]

# Połączenie wszystkich DataFrame's w jeden DataFrame
df = pd.concat(dane, ignore_index=True)

df
```

```python
# Stworzenie df1 jako pomniejszony Dataframe
df1 = df[['Number', 'Theme', 'Year', 'Set name']]
```

```python
df1
```

```python
# Stworzenie serii danych, gdzie indeksy to lata (kolumna "Year"), a wartości to liczba wystąpień zestawów w danym roku. 
lwr = df1['Year'].value_counts().sort_index()

# Różnice pomiędzy wartościami
roznice = np.diff(lwr.values)

# Najmniejsza pozytywna różnica
min_wartosc = np.min(np.abs(roznice))

# Największa różnica
max_wartosc = np.max(np.abs(roznice))
```

```python
# Tworzenie wykresu, gdzie osią OX są indexy z serii danych lwr oraz osią OY są wartości z lwr
plt.figure(figsize=(12, 7))
plt.plot(lwr.index, lwr.values, marker='o', linestyle='-', label='Zestawy', color='b', zorder=2)

# Linia trendu
# Pierwszym argumentem jest to oś OX (czyli lata)
# Drugim argumentem jest funkcja polyfit, która wykonuje dopasowanie wielomianu do danych, które mamy 
#     - Pierwszym argumentem jest oś OX, drugim jest oś OY, trzecim jest stopień wielomianu
#     - Jeżeli chcemy linie trendu, czyli funkcje liniową f(x)= ax+b to musimy dać stopień wielomianu równy 1 
#     - Wynik tej funkcji to lista dwuelementowa, gdzie elementem "0" jest współczynnik a, elementem "1" jest współczynnik b
plt.plot(lwr.index, np.polyfit(lwr.index, lwr.values, 1)[0] * lwr.index + np.polyfit(lwr.index, lwr.values, 1)[1],
         "-", label="Linia trendu", color='tomato', zorder=1)

# Siatka, legenda, nazwy osi
plt.grid(False)
plt.legend()
plt.title('Liczba zestawów wypuszczonych w poszczególnych latach')
plt.xlabel('Rok')
plt.ylabel('Liczba zestawów')

# Tworzenie własnej osi OX
plt.xticks(lwr.index, rotation=45)

# Dodawanie numerków nad punktami z liczbą zestawów

# tekst: Tekst, czyli liczba zestawów w danym roku.
# (lwr.index[i], txt): Współrzędne punktu, nad którym ma być umieszczony tekst. Oś x to rok, oś y to ilość zestawów.
# textcoords="offset points": Określa, że xytext (następny argument) to przesunięcie od punktu docelowego.
# xytext=(0, 10): Przesunięcie tekstu nad punktem. Tekst będzie umieszczony 10 punktów nad odpowiadającym punktem.
# ha='center': Wyrównanie tekstu w poziomie, w tym przypadku do środka.
    
for i, tekst in enumerate(lwr.values):
    plt.annotate(tekst, (lwr.index[i], tekst), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9, zorder=3)

# Strzałka skierowana pomiędzy dwie wartości z najmniejszą różnicą
plt.annotate(f'Najmniejsza różnica: {min_wartosc}',
             xy=(2019.5, 850),
             xytext=(2017.5, 750),
             arrowprops=dict(facecolor='yellow', shrink=0.05), zorder=4)

# Strzałka skierowana pomiędzy dwie wartości z największą różnicą
plt.annotate(f'Największa różnica: {max_wartosc}',
             xy=(2011.7, 650),
             xytext=(2013, 620),
             arrowprops=dict(facecolor='yellow', shrink=0.05), zorder=4)

plt.savefig('Liczba zestawów wypuszczonych w poszczególnych latach.pdf')

# Pokazanie wykresu
plt.show()
```
