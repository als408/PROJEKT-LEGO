```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('Brickset-Sets-2000.csv', sep = ',')
df2 = pd.read_csv('Brickset-Sets-2001.csv', sep = ',')
df3 = pd.read_csv('Brickset-Sets-2002.csv', sep = ',')
df4 = pd.read_csv('Brickset-Sets-2003.csv', sep = ',')
df5 = pd.read_csv('Brickset-Sets-2004.csv', sep = ',')
df6 = pd.read_csv('Brickset-Sets-2005.csv', sep = ',')
df7 = pd.read_csv('Brickset-Sets-2006.csv', sep = ',')
df8 = pd.read_csv('Brickset-Sets-2007.csv', sep = ',')
df9 = pd.read_csv('Brickset-Sets-2008.csv', sep = ',')
df10 = pd.read_csv('Brickset-Sets-2009.csv', sep = ',')
df11 = pd.read_csv('Brickset-Sets-2010.csv', sep = ',')
df12 = pd.read_csv('Brickset-Sets-2011.csv', sep = ',')
df13 = pd.read_csv('Brickset-Sets-2012.csv', sep = ',')
df14 = pd.read_csv('Brickset-Sets-2013.csv', sep = ',')
df15 = pd.read_csv('Brickset-Sets-2014.csv', sep = ',')
df16 = pd.read_csv('Brickset-Sets-2015.csv', sep = ',')
df17 = pd.read_csv('Brickset-Sets-2016.csv', sep = ',')
df18 = pd.read_csv('Brickset-Sets-2017.csv', sep = ',')
df19 = pd.read_csv('Brickset-Sets-2018.csv', sep = ',')
df20 = pd.read_csv('Brickset-Sets-2019.csv', sep = ',')
df21 = pd.read_csv('Brickset-Sets-2020.csv', sep = ',')
df22 = pd.read_csv('Brickset-Sets-2021.csv', sep = ',')
df23 = pd.read_csv('Brickset-Sets-2022.csv', sep = ',')
df24 = pd.read_csv('Brickset-Sets-2023.csv', sep = ',')

dane = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21, df22, df23, df24]

df = pd.concat(dane, ignore_index=True)

df
```

```python
df.drop(['Minifigs', 'RRP (GBP)', 'RRP (CAD)', 'RRP (EUR)', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 
         'Unnamed: 28', 'Unnamed: 29','Unnamed: 30', 'Unnamed: 31','Launch date', 'Exit date', 'EAN', 'UPC', 'Width', 
         'Height', 'Depth', 'Weight', 'Notes', 'Priority', 'Qty owned', 'Qty owned new', 'Qty owned used', 'Wanted', 'Qty wanted', 'Value new (USD)', 'Value used (USD)'], axis = 1, inplace = True)
df
# usunęłam niepotrzebne kolumny
```

```python
dfr = df.groupby('Year')['RRP (USD)'].mean()   
lata = dfr.index.tolist()
ceny = dfr.values.tolist()
# utworzyłam listę z latami i średnią ceną roczną
```

```python
indeks1 = ceny.index(max(ceny))
print(indeks1, max(ceny), lata[indeks1])
# wyznaczyłam: indeks najwyższej ceny w liście 'ceny', wartość tej ceny, rok w którym się pojawiła
```

```python
indeks2 = ceny.index(min(ceny))
print(indeks2, min(ceny), lata[indeks2])
# tu to samo tylko z najmniejszą ceną
```

```python
coefficients = np.polyfit(dfr.index, dfr.values, 1)
# ta zmienna tworzy wielomian stopnia 1, posługując się dfr.index jako x i dfr.values jako y
```

```python
fig, ax = plt.subplots(figsize = (15, 7))

plt.plot(dfr.index, dfr.values, color = 'red', label = 'Średnia cena zestawów LEGO w latach 2000 - 2023')
plt.plot(dfr.index, np.polyval(coefficients, dfr.index), color = 'gold', label = 'Linia trendu')
# np. polyval wykorzystuje współczynniki wielomianu (coefficients), które zostały wcześniej uzyskane 
# za pomocą np.polyfit(), do obliczenia wartości wielomianu dla odpowiadających punktów z dfr.index.

ax.set_xticks(np.arange(2000, 2024))
ax.set_yticks(np.arange(15, 55, step = 5))

ax.annotate('$20.35, 2001 r. \n Najniższa średnia roczna cena', xy = (2001, ceny[1]), xytext = (2004, 18),
            arrowprops = dict(arrowstyle = "->", connectionstyle = "angle3, angleA=0, angleB=90"))
# tu robię strzałkę, której współrzędne zależą od punktów na wykresie, za x biorę rok wystąpienia najniższej ceny,
# a za y jej wartość z listy (dzięki wyznaczonym indeksom), przy xytext ustawiałam współrzędne tak aby nie najeżdzały na wykres
# nie było ku temu żadnej reguły. Przy drugiej zrobiłam analogicznie

ax.annotate('$54.30, 2023 r. \n Najwyższa średnia roczna cena', xy = (2023, ceny[23]), xytext = (2016, 50),
            arrowprops = dict(arrowstyle = "->", connectionstyle = "angle3, angleA=0, angleB=90"))

plt.xlabel('Lata', fontsize = 14, fontname = 'Times New Roman')
plt.ylabel('Średnia roczna cena (USD)', fontsize = 14, fontname = 'Times New Roman')
# tytuły osi

plt.title('Czy średnia roczna cena zestawów LEGO wzrasta z roku na rok?', fontsize = 16, fontname = 'Times New Roman')
plt.legend()

plt.savefig('Czy średnia roczna cena zestawów LEGO wzrasta z roku na rok.pdf')
plt.show()
```
