```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import Normalize

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('Brickset-Sets-2000.csv', sep = ',')
df2 = pd.read_csv('Brickset-Sets-2001.csv', sep = ',')
df3 = pd.read_csv('Brickset-Sets-2002.csv', sep = ',')
df4 = pd.read_csv('Brickset-Sets-2003.csv', sep = ',')
df5 = pd.read_csv('Brickset-Sets-2004.csv', sep = ',')
df6 = pd.read_csv('Brickset-Sets-2005.csv', sep = ',')
df7 = pd.read_csv('Brickset-Sets-2006.csv', sep = ',')
df8 = pd.read_csv('Brickset-Sets-2007.csv', sep = ',')
df9 = pd.read_csv('Brickset-Sets-2008.csv', sep = ',')
df10 = pd.read_csv('Brickset-Sets-2009.csv', sep = ',')
df11 = pd.read_csv('Brickset-Sets-2010.csv', sep = ',')
df12 = pd.read_csv('Brickset-Sets-2011.csv', sep = ',')
df13 = pd.read_csv('Brickset-Sets-2012.csv', sep = ',')
df14 = pd.read_csv('Brickset-Sets-2013.csv', sep = ',')
df15 = pd.read_csv('Brickset-Sets-2014.csv', sep = ',')
df16 = pd.read_csv('Brickset-Sets-2015.csv', sep = ',')
df17 = pd.read_csv('Brickset-Sets-2016.csv', sep = ',')
df18 = pd.read_csv('Brickset-Sets-2017.csv', sep = ',')
df19 = pd.read_csv('Brickset-Sets-2018.csv', sep = ',')
df20 = pd.read_csv('Brickset-Sets-2019.csv', sep = ',')
df21 = pd.read_csv('Brickset-Sets-2020.csv', sep = ',')
df22 = pd.read_csv('Brickset-Sets-2021.csv', sep = ',')
df23 = pd.read_csv('Brickset-Sets-2022.csv', sep = ',')
df24 = pd.read_csv('Brickset-Sets-2023.csv', sep = ',')

dane = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21, df22, df23, df24]

df = pd.concat(dane, ignore_index=True)

df
```

```python
df.drop(['Minifigs', 'RRP (GBP)', 'RRP (CAD)', 'RRP (EUR)', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 
         'Unnamed: 28', 'Unnamed: 29','Unnamed: 30', 'Unnamed: 31','Launch date', 'Exit date', 'EAN', 'UPC', 'Width', 
         'Height', 'Depth', 'Weight', 'Notes', 'Priority', 'Qty owned', 'Qty owned new', 'Qty owned used', 'Wanted', 'Qty wanted', 'Value new (USD)', 'Value used (USD)'], axis = 1, inplace = True)
df
# wyrzuciłam niepotrzebne kolumny
```

```python
dfr = df.groupby('Year')['Pieces'].mean()
lata = dfr.index.tolist()
# wyznaczyłam dfr tylko po to, żeby móc wyznaczyć sobie listę lat
```

```python
rekordy = []
# tworzę listę, w której będą rekordy ilości klocków z każdego roku

for i in range(2000, 2024):  
    pi = df[df['Year'] == i]   # tu wyznaczam sobie fragment dataframu dla konkretnego roku
    li = pi['Pieces'].tolist()   # robię listę ilości klocków z zestawów z tego roku
    rekordy.append(max(li)) # i największą wartość wrzucam do listy
```

```python
rekordy
```

```python
dfmin = df[(df['Year'] == 2001) & (df['Pieces'] == 1868)]
dfmin
# po wyswietleniu wykresu wybrałam rok i ilość klocków gdzie było ich najmniej, aby dowiedzieć się jaki zestaw miał taki wynik
```

```python
dfmax = df[(df['Year'] == 2021) & (df['Pieces'] == 11695)]
dfmax
# tu to samo tylko z największą ilością
```

```python
fig, ax = plt.subplots(figsize = (15, 10))

ax.barh(lata, list(reversed(rekordy)), tick_label = list(reversed(lata)), color = plt.cm.autumn(np.linspace(0, 1, len(rekordy))))
# rekordy i dalej lata dałam na reversed, aby się wyświetlały się chronologicznie
# mapa kolorów nie jest dostosowana do odpowiednich wartości bo dla większości przyjmowała ten sam koloru ze względu na
# rozstrzał w ilości klocków. 
# np. linspace tworzy równomiernie rozłożony ciąg liczb od 0 do 1 o długości równej liczbie 
# rekordów (len(rekordy)). Te liczby są używane do określenia stopni koloru w mapie 'autumn'.

ax.set_xlim(0, 13000)
ax.set_xticks([])   # to ustawiłam, żeby nie wyświetlały się wartości x na dole wykresu

ax.text(4500, 2020, 'Zestaw z najmniejszym rekordem 1868 klocków (2001r.) - "Star Wars - Ultimate Collector Series - Darth Maul" \n'
'Zestaw z największym rekordem 11695 klocków (2021r.) - "Art - Miscellaneous - World Map"', 
        fontsize = 12, fontname = 'Times New Roman')
# współrzędne tego tekstu ustawiałam losowo aż mi pasowało \n dodało mi drugą linijkę tego tekstu

for i, liczba in zip(lata, list(reversed(rekordy))):
    ax.text(liczba, i, str(liczba), ha='left', va='center', color='black', fontsize = 11, fontname = 'Times New Roman')
# tu stworzyłam pętlę która dodała mi teksty przy słupkach. Po kolei była wartość x, y, tekst czyli ilość klocków 
# i umiejscowienie jako 'left', aby znajdowało się obok słupka

plt.xlabel('Ilość elementów w zestawie', fontsize = 14, fontname = 'Times New Roman')
# to tytuł osi X
plt.ylabel('Lata', fontsize = 14, fontname = 'Times New Roman')
# a tu Y

plt.title('Rekordowe ilości elementów w zestawach w latach 2000 - 2023', fontsize = 16, fontname = 'Times New Roman')

plt.savefig('Rekordowe ilości elementów w zestawach w latach 2000 - 2023.pdf')
plt.show()
```
